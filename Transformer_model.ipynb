```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import label_binarize
from itertools import cycle
import random
import pdfkit
from io import BytesIO
import base64
import os
from tensorflow.keras.layers import MultiHeadAttention, LayerNormalization, Dropout

# Set the Google Drive save path
save_path = '/content/drive/MyDrive/BDA/Model_Evaluation_Reports'
if not os.path.exists(save_path):
    os.makedirs(save_path)
def load_and_preprocess_data(file_path, min_seq_length=50):
    data = pd.read_csv(file_path)
    data.dropna(subset=['seq'], inplace=True)
    data['seq'] = data['seq'].str.upper().str.replace('N', '')
    data = data[data['seq'].str.len() >= min_seq_length]
    data.drop_duplicates(subset=['seq'], inplace=True)
    sequences = data['seq'].astype(str).tolist()
    labels = data['id'].tolist()
    return sequences, labels

# K-mer encoding with padding
def kmer_encoding(sequences, k=3, max_len=500):
    kmers = []
    for seq in sequences:
        kmer = [seq[i:i+k] for i in range(len(seq) - k + 1)]
        kmers.append(' '.join(kmer))
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(kmers)
    sequences_encoded = tokenizer.texts_to_sequences(kmers)
    sequences_padded = pad_sequences(sequences_encoded, maxlen=max_len, padding='post')
    vocab_size = len(tokenizer.word_index) + 1
    return sequences_padded, vocab_size

# Build a basic Transformer model
def build_basic_transformer_model(vocab_size, input_length, num_classes):
    # Input and Embedding layers with minimal embedding size
    inputs = layers.Input(shape=(input_length,))
    embedding_layer = layers.Embedding(input_dim=vocab_size, output_dim=8)(inputs)  # Very low embedding dimension

    # Single Attention Head with no skip connections or additional normalization layers
    attention = layers.MultiHeadAttention(num_heads=1, key_dim=4)(embedding_layer, embedding_layer)  # One head, low key dimension

    # Pooling and Output
    pooled = layers.GlobalMaxPooling1D()(attention)  # Directly pool the attention output
    outputs = layers.Dense(num_classes, activation='softmax')(pooled)  # Simple output layer

    # Compile the model with basic settings
    model = models.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Helper to convert figures to base64 for inline HTML images
def fig_to_base64(fig):
    img = BytesIO()
    fig.savefig(img, format="png", bbox_inches="tight")
    plt.close(fig)  # Close figure to prevent memory leaks
    return base64.b64encode(img.getvalue()).decode()

# Function to save HTML and PDF report
def save_report_to_html_pdf(html_content):
    html_file_path = os.path.join(save_path, "Consolidated_Evaluation_Report_Transformer.html")
    pdf_file_path = os.path.join(save_path, "Consolidated_Evaluation_Report_Transformer.pdf")

    # Save HTML content to file
    with open(html_file_path, "w") as file:
        file.write(html_content)

    # Convert HTML file to PDF
    pdfkit.from_file(html_file_path, pdf_file_path)

# Train and evaluate the model with different data sizes and test ratios
def train_and_evaluate(file_path):
    sequences, labels = load_and_preprocess_data(file_path)
    total_samples = len(sequences)
    data_size_percentages = [0.1, 0.25, 0.5, 0.75, 1.0]
    test_ratios = [0.2, 0.3, 0.4, 0.5]

    html_content = "<h1>Consolidated Transformer Model Evaluation Report</h1>"

    for percentage in data_size_percentages:
        data_size = int(total_samples * percentage)
        sampled_idx = random.sample(range(total_samples), data_size)
        sampled_sequences = [sequences[i] for i in sampled_idx]
        sampled_labels = [labels[i] for i in sampled_idx]

        X_kmer, vocab_size = kmer_encoding(sampled_sequences)
        unique_labels, labels_encoded = np.unique(sampled_labels, return_inverse=True)
        class_names = unique_labels
        num_classes = len(unique_labels)

        for test_ratio in test_ratios:
            title_suffix = f"{data_size} samples ({int(percentage * 100)}% data), Test Ratio: {int(test_ratio * 100)}%"
            print(f"\nTraining with {title_suffix}")
            html_content += f"<h2>Data Size: {data_size} samples, Test Ratio: {test_ratio}</h2>"

            X_train, X_test, y_train, y_test = train_test_split(X_kmer, labels_encoded, test_size=test_ratio, random_state=42)

            y_train_one_hot = to_categorical(y_train, num_classes=num_classes)
            y_test_one_hot = to_categorical(y_test, num_classes=num_classes)

            model = build_basic_transformer_model(vocab_size, X_kmer.shape[1], num_classes)
            early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

            history = model.fit(X_train, y_train_one_hot,
                                validation_data=(X_test, y_test_one_hot),
                                epochs=40,
                                batch_size=32,
                                callbacks=[early_stopping],
                                verbose=1)

            y_pred = np.argmax(model.predict(X_test), axis=1)
            y_pred_prob = model.predict(X_test)
            metrics_data = {
                "accuracy": accuracy_score(y_test, y_pred),
                "precision": precision_score(y_test, y_pred, average='macro',  zero_division=1),
                "recall": recall_score(y_test, y_pred, average='macro'),
                "f1_score": f1_score(y_test, y_pred, average='macro')
            }

            print(f"\nEvaluation Metrics for {title_suffix}")
            for key, value in metrics_data.items():
                print(f"{key.capitalize()}: {value:.4f}")
            html_content += "<ul>"
            for key, value in metrics_data.items():
                html_content += f"<li>{key.capitalize()}: {value:.4f}</li>"
            html_content += "</ul>"

            # Confusion Matrix
            cm = confusion_matrix(y_test, y_pred)
            fig, ax = plt.subplots(figsize=(10, 7))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, ax=ax)
            plt.xlabel('Predicted')
            plt.ylabel('True')
            plt.title(f'Confusion Matrix for {title_suffix}')
            plt.show()
            html_content += f'<h3>Confusion Matrix for {title_suffix}</h3><img src="data:image/png;base64,{fig_to_base64(fig)}" />'

            # ROC and Precision-Recall Curves
            roc_fig, pr_fig = plot_metrics(y_test, y_pred, y_pred_prob, num_classes, class_names, title_suffix)
            plt.show()

            html_content += f'<h3>ROC Curve for {title_suffix}</h3><img src="data:image/png;base64,{fig_to_base64(roc_fig)}" />'
            html_content += f'<h3>Precision-Recall Curve for {title_suffix}</h3><img src="data:image/png;base64,{fig_to_base64(pr_fig)}" />'

            # Training History
            epochs = range(1, len(history.history['loss']) + 1)

            # Model Accuracy Plot
            fig, ax = plt.subplots()
            ax.plot(epochs, history.history['accuracy'], label='Train Accuracy')
            ax.plot(epochs, history.history['val_accuracy'], label='Validation Accuracy')
            plt.title(f'Model Accuracy for {title_suffix}')
            plt.xlabel('Epoch')
            plt.ylabel('Accuracy')
            plt.legend()
            plt.show()
            html_content += f'<h3>Model Accuracy for {title_suffix}</h3><img src="data:image/png;base64,{fig_to_base64(fig)}" />'

            # Model Loss Plot
            fig, ax = plt.subplots()
            ax.plot(epochs, history.history['loss'], label='Train Loss')
            ax.plot(epochs, history.history['val_loss'], label='Validation Loss')
            plt.title(f'Model Loss for {title_suffix}')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            plt.show()
            html_content += f'<h3>Model Loss for {title_suffix}</h3><img src="data:image/png;base64,{fig_to_base64(fig)}" />'

    save_report_to_html_pdf(html_content)

# Plot metrics (ROC and Precision-Recall curves)
def plot_metrics(y_test, y_pred, y_pred_prob, num_classes, class_names, title_suffix):
    y_test_bin = label_binarize(y_test, classes=list(range(num_classes)))

    fpr, tpr, roc_auc = {}, {}, {}
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    fig, ax = plt.subplots()
    colors = cycle(['blue', 'red', 'green', 'orange', 'purple', 'pink'])
    for i, color in zip(range(num_classes), colors):
        ax.plot(fpr[i], tpr[i], color=color, lw=2, label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve for {title_suffix}')
    ax.legend(loc="lower right")
    roc_fig = fig

    fig, ax = plt.subplots()
    for i, color in zip(range(num_classes), colors):
        precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_pred_prob[:, i])
        pr_auc = auc(recall, precision)
        ax.plot(recall, precision, color=color, lw=2, label=f'{class_names[i]} (AUC = {pr_auc:.2f})')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    ax.set_title(f'Precision-Recall Curve for {title_suffix}')
    ax.legend(loc="upper left")
    pr_fig = fig

    return roc_fig, pr_fig

if __name__ == "__main__":
    file_path = '/content/drive/MyDrive/BDA/merge.csv'
    train_and_evaluate(file_path)
```
